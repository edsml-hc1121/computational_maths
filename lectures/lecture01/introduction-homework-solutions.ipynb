## Homework - Analysing the classical form of the logistic map

Recreate the analysis of the logistic map conducted in class for the initial form of the map, i.e. with the following code for RHS function.

Specifically plot the RHS; compute analytically and plot the fixed points; investigate the sensitivity of the results to different $x_0$ and $r$ values; and plot the bifurcation diagam.

````python
def logistic_map(r, x):
    """A function giving the RHS of the discrete logistic map

    Parameters
    ----------
    r : float
        Growth rate parameter
    x : float
        Current population $x_n$, value from 0 to 1

    Returns
    -------
    float
        The updated population $x_{n+1} = r x_n (1-x_n)$
    """
    return r*x*(1-x)
````

### Solution - Classical form of the logistic map

````python
# Define a parameter value r
r = 2

# Possible population values (x) are between zero and one,
# create a vector of possible values for plotting purposes
x = np.linspace(0, 1, 100)
# 100 points should be enough to get a nice smooth line, but you 
# could increase this.
# You could also have used np.arange

# Compute the corresponding logistic_map values, as we'll be
# plotting these, let's just use variable y to store these
y = logistic_map(r, x)

# Set up a figure for plotting
fig = plt.figure(figsize=(8, 6))
ax1 = fig.add_subplot(111)

# plot the data
ax1.plot(x,y,'b')

# add axes labels and a title, play with the fontsizes to make it look nice
ax1.set_title('The logistic map ($r={0:.3f}$)'.format(r), fontsize=16)
ax1.set_xlabel('$x$', fontsize=14)
ax1.set_ylabel('$rx(1-x)$', fontsize=14)
# add a grid
ax1.grid(True)
````

#### Fixed points

As we did in class for the other form of the logistic map, let's see if our problem has any *fixed points*, i.e. a population that if reached then doesn't vary in future generations. 

(Note that as discussed in the lecuture for this model $x=1$ is not a fixed point, this state implies that we will exhaust all our resources in that generation and so in the next discrete step the population goes to zero as everyone dies and no-one is born!)

Mathematically, for our problem, we write this as

\begin{align*}
x_{n+1} &= x_{n} \\
\iff rx_n(1-x_n) &= x_n
\end{align*}

What two values of $x_n$ does this correspond to?

We can work this our mathematically ... 

[$x_n=0$ and $ r(1-x_n)=1 \implies x_n = 1-1/r$]

and/or we can just plot the expressions on either side of this last equation to figure this out (plotting both sides gives us two lines of course, one of which is just $y=x$, the other is our logistic map RHS we've already plotted).

````python
# choose r = 3 this time
r = 3
x = np.linspace(0, 1, 100)
y = logistic_map(r, x)

# set up a figure and axis
fig = plt.figure(figsize=(8, 6))
ax1 = fig.add_subplot(111)

# plot logistic map
ax1.plot(x,y,'b',label=r'$y = rx(1-x)$')
ax1.set_title('The logistic map ($r={0:.2f}$)'.format(r), fontsize=16)
ax1.set_xlabel('$x$', fontsize=14)
# add a grid
ax1.grid(True)

# plot y = x line to find fixed points
# since at fixed points the next generation's population is equal to the current
ax1.plot(x,x,'k',label=r'$y = x$')

# plot the two intersection points I worked out mathematically
ax1.plot(0, logistic_map(r,0), 'ro')
ax1.plot(1-1./r, logistic_map(r,1-1./r), 'ro', label='intersection points')
ax1.legend(loc='best', fontsize=14)
````

What does this plot tell us?

Firstly that we do indeed have two intersection points, assuming $r$ is greater than one. 

Ignoring the fixed point at (0,0) for now [a zero population in one generation will produce no offspring so remain at zero - this isn't a very interesting fixed point], 

if our population is below the $x$ value that corresponds to the non-zero intersection value it will clearly grow in the next generation - remember for a population given by a value on the $x$ axis, the blue line tells us the next generation's population size, with the black line the current generation size.



But if it is above that value, the next generation will be lower. This is a stable fixed point - if the population is slightly displaced from the fixed point, it will move back to the fixed point.

We can plot this evolution of population sizes in what is sometimes referred to as a [cobweb plot](https://en.wikipedia.org/wiki/Cobweb_plot).

````python
# first define a function to do this
def plot_logistic_map_convergence(r, x0, iterations=10, ax=None):
    """Function to plot the iterations that the logistic map undergoes
    on axes ax (if given)

    Parameters
    ----------
    r : float
        Growth rate parameter
    x0 : float
        Initial population, value from 0 to 1
    iterations : float, optional
        The number of iterations to compute
    ax : matplotlib axes, optional
        The axes to plot on
    """
    # start by plotting the underlying curves from above
    x = np.linspace(0, 1, 100)
    y = logistic_map(r, x)
    ax.plot(x,y,'b',label=r'$y = rx(1-x)$')
    ax.plot(x,x,'k',label=r'$y = x$')
    ax.set_title('The logistic map ($r={0:.2f}$)'.format(r), fontsize=16)
    ax.set_xlabel('$x$', fontsize=14)
    ax.grid(True)
    ax.legend(loc='best', fontsize=14)
    # plot our starting point
    ax.plot(x0, x0, 'go') 
    # now iterate upto "iterations" and plot the path taken in (x,y) space
    x = x0
    for it in range(iterations):
        y = logistic_map(r,x)
        # plot some lines to indicate map progression
        ax.plot([x,x],[x,y],'k--',linewidth=1)
        ax.plot([x,y],[y,y],'k--',linewidth=1)
        # add a dot of increasing boldness
        ax.plot(x, y, 'ro', alpha=(it+1)/iterations)
        x=y
````

````python
# now set up a figure and run the function
fig = plt.figure(figsize=(8, 6))
ax1 = fig.add_subplot(111)

r = 1.9
x0 = 0.2
plot_logistic_map_convergence(r, x0, 10, ax1)
````

If it's not clear what this diagram is telling us:

1. We start from an $x$ value - call it $x_0$ and plot it on the $y=x$ line of the plot - it's the green dot (this is the `x = x0` line in the code above).


2. Use the logistic map to find the next value which is given by the blue line - the vertical black dashed line up from the green dot indicates this operation (the `y = logistic_map(r,x)` line in the code above computes the value, and `ax.plot(x, y, ... ` plots it).


3. Now we need to store this new value for use in the next iteration, i.e. we effectively set $x_1 = y$. We don't actually store all the iterations in this implementation, e.g. in a vector of length the number of iterations we are going to perform (we could easily do this), so instead we just overwrite our $x$ variable (`x=y` in the code above) for use in the next iteration. In the diagram this operation is indicated by the horizontal line that takes us back to the $y=x$ curve.


4. And we repeat. To help identify the iterations we change the opacity of the red dots - this will come in more useful shortly!

For this example, how would we get the system to converge to the fixed point at zero? I.e, under what conditions would the unstable fixed point at zero turn into a stable fixed point?

Hint: Think about choice of parameter $r$ rather than a starting location.

Second hint: try a value of $r$ less than unity!

As discussed in class, there are two parameters we can play with here - the initial condition $x_0$ and the parameter $r$.

Thinking about this as a generic numerical algorithm, it's vitally important that we understand/appreciate the sensitivity of our results to **both** of these parameters for the reasons explained in class.

So, for the $r$ parameter value considered here at least, we always end up at the same place.

#### Bifurcation diagram

So if we vary $r$ we can get quite different behaviour - both the intermediate steps as well as where we end up in the long run. 

Let's plot the ultimate behaviour as a function of the parameter $r$ in a so-called [bifurcation diagram](https://en.wikipedia.org/wiki/Bifurcation_diagram).

NB. this implementation is [*vectorised*](https://en.wikipedia.org/wiki/Array_programming). See also <https://www.geeksforgeeks.org/vectorization-in-python/>.  If we looped over all the $r$ values this code would run much slower.

````python
# consider N different r values - the bigger the N we select, the longer it will take!
N = 10000
r = np.linspace(2.0, 4.0, N)
# corresponding N initial x values
x = 0.1*np.ones(N)
# we're interested in long term limiting behaviour, so do quite a few iterations of the map 
iterations = 500
# but only plot the final 100 values we hit
values_to_plot = 100
# compute the logistic map dynamics for each r value, 
# and plot the final x values we hit for each r
fig = plt.figure(figsize=(14, 6))
ax1 = fig.add_subplot(111)
for it in range(iterations):
    x = logistic_map(r,x)
    if it>(iterations-values_to_plot):
        ax1.plot(r,x,'k,',alpha=0.1)
ax1.set_title('Bifurcation diagram for the logistic map', fontsize=16)
ax1.set_xlabel('$r$', fontsize=14)
ax1.set_ylabel('Eventual Population', fontsize=14)
````

The code starts at $x_0=0.1$ for each $r$ values, iterates the map 500 times, and plots on the $y$ axis the final 100 populations sizes (the first 400 could be called a "spin up" period). 

So on the plot, looking up from an $r$, distinct values tells us the map converges to a value or values, whereas a "cloud" of points indicates that these 100 values are jumping around.

<br>

As $r$ increases from 2 and passes 3, the stable fixed point becomes an unstable fixed point and two "arms" appear - that is the "mode 2" solution we saw earlier, where the population jumps between these two points. As we increase $r$ further, the system eventually becomes chaotic, until it completely diverges after $r = 4$ .

## Homework - Perturbations in the logistic map

As mentioned in the lecture, compute and plot how the value of an initial perturbation in starting location varies with iteration number.

For example, you could write a function that takes two starting values (one the perturbation of the other) and plot how the difference between the two changes with iteration number.

Investigate how this changes with different starting points, perturbation size and choice of $r$.

### Solution - Perturbations in the logistic map

Start with the code and example from the lecture.

````python
def logistic_map(r, x):
    return r*x*(1-x)

def plot_logistic_map_convergence(r, x0, iterations=10, ax=None):
    """Function to plot the iterations that the logistic map undergoes
    on axes ax (if given)

    Parameters
    ----------
    r : float
        Growth rate parameter
    x0 : float
        Initial population, value from 0 to 1
    iterations : float, optional
        The number of iterations to compute
    ax : matplotlib axes, optional
        The axes to plot on
    """
    # start by plotting the underlying curves from above
    x = np.linspace(0, 1, 100)
    y = logistic_map(r, x)
    ax.plot(x,y,'b',label=r'$y = rx(1-x)$')
    ax.plot(x,x,'k',label=r'$y = x$')
    ax.set_title('The logistic map ($r={0:.2f}$)'.format(r), fontsize=16)
    ax.set_xlabel('$x$', fontsize=14)
    ax.grid(True)
    ax.legend(loc='best', fontsize=14)
    # plot our starting point
    ax.plot(x0, x0, 'go') 
    # now iterate up to "iterations" and plot the path taken in (x,y) space
    x = x0
    for it in range(iterations):
        y = logistic_map(r,x)
        # plot some lines to indicate map progression
        ax.plot([x,x],[x,y],'k--',linewidth=1)
        ax.plot([x,y],[y,y],'k--',linewidth=1)
        # add a dot of increasing boldness
        ax.plot(x, y, 'ro', alpha=(it+1)/iterations)
        x=y
        
# Let's have two plots side by side
fig, axs = plt.subplots(1, 2, figsize=(14, 7))

r=1.9
x0 = 0.7
plot_logistic_map_convergence(r, x0, 10, axs[0])
pert = 0.05
plot_logistic_map_convergence(r, x0 + pert, 10, axs[1])
````

Now write a new function which takes two starting values and outputs the difference.

````python
def compute_logistic_map_perturbations(r, x1_0, x2_0, iterations=10):
    """Function to compute size of perturbation for every iteration.

    Parameters
    ----------
    r : float
        Growth rate parameter
    x1_0 : float
        Initial population, value from 0 to 1
    x2_0 : float
        Perturbed initial population
    iterations : float, optional
        The number of iterations to compute

    Returns
    -------
    perts : numpy.ndarray
        Array of perturbation value at each iteration
    """
    # we want both our initial conditions to be between zero and one so let's check they both are
    assert (0<=x1_0<=1)
    assert (0<=x2_0<=1)
        
    # set up array and initial values for unperturbed and perturbed case
    perts = np.zeros(iterations+1)
    perts[0] = x2_0 - x1_0  
    x1 = x1_0
    x2 = x2_0
    # loop through iterations and compute difference between unperturbed and perturbed
    for it in range(iterations):
        y1 = logistic_map(r,x1)
        y2 = logistic_map(r,x2)
        perts[it+1] = y2 - y1
        # update population for next iteration
        x1 = y1
        x2 = y2
    return perts
````

Run it and plot the output for some difference starting values, perturbation sizes, and $r$ values.

````python
# plot perturbation values by iteration for r = 1.9

r = 1.9
x1_0 = 0.7
pert = 0.01
x2_0 = x1_0 + pert
perts = compute_logistic_map_perturbations(r, x1_0, x2_0, 100)

fig = plt.figure(figsize=(6, 6))
ax1 = plt.subplot(111)

ax1.set_title('Variation in perturbation with iteration', fontsize=16)
ax1.plot(range(len(perts)), perts, 'b.-')
ax1.set_xlabel('$n$', fontsize=16)
ax1.set_ylabel('$x_n$', fontsize=16)
````

We saw an [animated gif from Wikipedia](https://en.wikipedia.org/wiki/Taylor_series#/media/File:Exp_series.gif) at the start of the lecture for how the Taylor series approximation to the exponential function, around the point $x=0$, more closely resembles the exponential function over the domain $[-3,3]$ as more terms are included in the expansion:

<img src="https://upload.wikimedia.org/wikipedia/commons/6/62/Exp_series.gif" width=200x>


Write a script which computes and plots this approximation for increasing numbers of terms in the expansion (essentially recreating the individual frames that make up that animation).

In addition, plot the convergence of the RMS ("root mean square") error (you may need to look up what this is defined to be - we will go over in detail in L4) computed using 1000 points over the interval $[-3,3]$ against the number of terms included in the expansion on the x-axis.

Hint: since the derivative of $\exp(x)$ is also $\exp(x)$, you can re-use a single function for all the derivatives you require in the Taylor series expansion.  You won't be able to do this for other functions of course!

### Solution - Taylor series

````python
from math import factorial

fig = plt.figure(figsize=(14, 7))
ax1 = plt.subplot(121)

ax1.set_title('Taylor series approximation', fontsize=16)
ax1.set_xlabel('$x$', fontsize=16)
ax1.set_ylabel('$f(x)$', fontsize=16)

# exp is a convenient function to use as an example since all the derivatives are the same: df/dx = f, etc

def f(x):
    return np.exp(x)

# define our x for plotting purposes
x = np.linspace(-3., 3., 1000)

# plot exact function
ax1.plot(x, f(x), 'k', lw=6, label='Exact function')

# expansion point
x0 = 0.

taylor_series_expansion = np.zeros_like(x)
N = 10
errors = np.zeros(N)
# plot Taylor series truncated at increasing numbers of terms
for i in range(N):
    # rather than functions for higher order derivs, for exp example we can just keep pointing at f(.)
    taylor_series_expansion += f(x0) * (x - x0)**i / factorial(i)
    # compute the error in our approximation.  
    errors[i] = np.linalg.norm( taylor_series_expansion - f(x) ) / np.sqrt(len(x))
    # QUESTION - why have I divided by square root of length of x here ???????
    # plot the approximation as a function of x
    ax1.plot(x, taylor_series_expansion, label='Number of terms = {0:1d}'.format(i))

ax1.legend(loc='best', fontsize=10)

# add a second plot for the errors
ax2 = plt.subplot(122)
ax2.plot(range(N), errors, 'b.-')
ax2.set_xlabel('Number of terms', fontsize=16)
ax2.set_ylabel('Error', fontsize=16)
ax2.set_title('Error in Taylor series expansion\n as function of number of terms', fontsize=16);
````

So for this function we can see that over the interval considered we can get pretty much perfect agreement as long as we retain enough terms in the Taylor series expansion.

## Homework - Taylor series (convergence)

Extending the example from the previous question.

Fix the number of terms in your Taylor series, i.e. truncate the series at the same term, and investigate how the error in your approximation decreases as you reduce how far you evaluate the approximation away from the expansion location.

That is, plot

$$| f(x_0 + h) - T(x_0 + h) |$$

i.e. the error at a single location, as a function of $h$, where $f$ is the true function (exp) and $T$ is your truncated Taylor series approximation.

Plot using a log-log plot - what slope line do you expect?

### Solution - Taylor series (convergence)

````python
from math import factorial

fig = plt.figure(figsize=(7, 7))
ax1 = plt.subplot(111)

ax1.set_title('Taylor series approximation', fontsize=16)
ax1.set_xlabel('$h$', fontsize=16)
ax1.set_ylabel('Error', fontsize=16)

# exp is a convenient function to use as an example since all the derivatives are the same: df/dx = f, etc

def f(x):
    return np.exp(x)


def taylor(f, order, x):
    """Function to compute the taylor series approximation.

    Parameters
    ----------
    f : function
        Input function to be approximated by the taylor series
    order : integer
        Number of terms
    x : float
        The point the taylor series is about

    Returns
    -------
    taylor_series_expansion : numpy.ndarray
        Array of taylor series approximation at each point
    """
    taylor_series_expansion = np.zeros_like(x)
    for i in range(order):
        taylor_series_expansion += f(x0) * (x - x0)**i / factorial(i)    
    return taylor_series_expansion


# expansion point
x0 = 0.

# define some increments from x0 to test
N = 6
h = np.pi*(2.**(-1*np.arange(N)))
# above creates a numpy array with first value pi, then pi/2, then pi/4, ...

# let's plot the errors with a choice for "order" using a log-log plot
order = 4
errors = np.abs( f(x0+h) - taylor(f,order,x0+h) )

ax1.loglog(h, errors, 'b.-', label = 'error')


# Compute the slope of the line and add the corresponding straight line to show how
# well our error plot agrees
start_fit = 0

line_fit1 = np.polyfit(np.log(h[start_fit:]), np.log(errors[start_fit:]), 1)
ax1.loglog(h[start_fit:], np.exp(line_fit1[1]) * h[start_fit:]**(line_fit1[0]), 'k-', label = 'slope: {:.2f}'.format(line_fit1[0]))

ax1.legend(loc='best', fontsize=14)
````

We observe a slope that is approximately the same as the `order` we set in the code.
You can vary this to confirm it holds for an integer value.


We will explain a bit better why this is in L4.

Note that our code agreeing with what the theory tells us should be the error is an example of what is called "code verification".

Repeat the previous exercise, but now consider the function $f(x) = \sin(x)$ for $x\in[-\pi/2,\pi/2]$ and an expansion point of $x_0 = 0$.

### Solution - Taylor series (a different function to class)

````python
from math import factorial

fig = plt.figure(figsize=(14, 7))
ax1 = plt.subplot(121)

ax1.set_title('Taylor series approximation', fontsize=16)
ax1.set_xlabel('$x$', fontsize=16)
ax1.set_ylabel('$f(x)$', fontsize=16)

# now I need to do some more work to generate a function that returns the derivatives,
# first we note that the derivatives of sin take 4 values before they repeat:
# cos, -sin, -cos, sin,  and then cos, ... again
def f(x, deriv):
    """ 
    Function to compute f(x)=sin(x) and its derivatives.

    Parameters
    ----------
    x : float or ndarray
        The point the taylor series is about
    deriv : int
        Order of the derivative

    Returns
    -------
    float or ndarray
        Function or derivative value at each point
    """
    if deriv%4 == 0:
        return np.sin(x)
    elif deriv%4 == 1: 
        return np.cos(x)
    elif deriv%4 == 2:
        return -np.sin(x)
    elif deriv%4 == 3:
        return -np.cos(x)


# define our x for plotting purposes
x = np.linspace(-np.pi/2, np.pi/2, 1000)

# plot exact function
ax1.plot(x, f(x,0), 'k', lw=3, label='Exact function')

# expansion point
x0 = 0.

taylor_series_expansion = np.zeros_like(x)
N = 10
errors = np.zeros(N)
# plot Taylor series truncated at increasing numbers of terms
for i in range(N):
    taylor_series_expansion += f(x0,i) * (x - x0)**i / factorial(i)
    errors[i] = np.linalg.norm( taylor_series_expansion - f(x,0) ) / np.sqrt(len(x))
    ax1.plot(x, taylor_series_expansion, label='Number of terms = {0:1d}'.format(i))

ax1.legend(loc='best', fontsize=10)
ax2 = plt.subplot(122)
ax2.plot(range(N), errors, 'b.-')
ax2.set_xlabel('Number of terms', fontsize=16)
ax2.set_ylabel('Error', fontsize=16)
ax2.set_title('Error in Taylor series expansion as function of number of terms', fontsize=16);
````

## Homework - Big O notation and large/small limits [*]

Consider the example from class

$$g(y) = 3y^2 - y^3 + 9y^4$$

Use plots to convince yourself that

$$g(y) = \mathcal{O}(y^2) \quad\text{as}\quad y\rightarrow 0$$

and

$$g(y) = \mathcal{O}(y^4) \quad\text{as}\quad y\rightarrow \infty$$

i.e. plot the function $g(y)$, as well as functions $Cy^2$ and $Cy^4$ to convince yourself that, respectively, as long as you look at small or large enough values of $y$ these do indeed bound $g$. 

</div>

### Solution - Big O notation

````python
def g(y):
    return 3*y**2 - y**3 + 9*y**4

# let's see if a quadratic function bounds g
def g_bound_small(y):
    return 13*y**2

fig = plt.figure(figsize=(12, 6))
ax1 = plt.subplot(121)
ax2 = plt.subplot(122)


# plot from y=0 to y=10
y = np.linspace(0, 10, 100)

# plot exact function
ax1.plot(y, g(y), 'k', lw=1, label='Exact function')
ax1.plot(y, g_bound_small(y), 'b', lw=1, label='Limiting behaviour')
ax1.legend(loc='best', fontsize=14)
ax1.set_title('Big O example', fontsize=16)
ax1.set_xlabel('$y$', fontsize=16)
ax1.set_ylabel('$g(y)$', fontsize=16)


# zoom in to y=0 to y=1.3 to examine behavior as y approaches 0 more closely
y = np.linspace(0, 1.3, 100)

# plot exact function
ax2.plot(y, g(y), 'k', lw=1, label='Exact function')
ax2.plot(y, g_bound_small(y), 'b', lw=1, label='Limiting behaviour')
ax2.legend(loc='best', fontsize=14)
ax2.set_title('Big O example (zoom)', fontsize=16)
ax2.set_xlabel('$y$', fontsize=16)
ax2.set_ylabel('$g(y)$', fontsize=16)
````

On the left we see that the blue line is NOT bounding our function (the black line). This is because we are not looking at small enough $y$ values to be in the "asymptotic limit"

This is because we are interested in the *limiting behaviour* - the figure on the right shows that for small enough $y$ values the blue line does indeed bound the black line.  Of course by changing the constant $C$ we could shift the location of the point at which the bounding happen, but the constant is not the important part, it's the power of the independent variable (here $y$) that's important.

````python
def g(y):
    return 3*y**2 - y**3 + 9*y**4

# let's see if a quartic function bounds g
def g_bound_large(y):
    return 13*y**4

fig = plt.figure(figsize=(12, 6))
ax1 = plt.subplot(121)
ax2 = plt.subplot(122)


# plot for smaller y values
y = np.linspace(0, 0.9, 100)

# plot exact function
ax1.plot(y, g(y), 'k', lw=1, label='Exact function')
ax1.plot(y, g_bound_large(y), 'b', lw=1, label='Limiting behaviour')
ax1.legend(loc='best', fontsize=14)
ax1.set_title('Big O example (smaller $y$ values)', fontsize=16)
ax1.set_xlabel('$y$', fontsize=16)
ax1.set_ylabel('$g(y)$', fontsize=16)


# plot for larger y values
y = np.linspace(0, 10, 100)

# plot exact function
ax2.plot(y, g(y), 'k', lw=1, label='Exact function')
ax2.plot(y, g_bound_large(y), 'b', lw=1, label='Limiting behaviour')
ax2.legend(loc='best', fontsize=14)
ax2.set_title('Big O example  (larger $y$ values)', fontsize=16)
ax2.set_xlabel('$y$', fontsize=16)
ax2.set_ylabel('$g(y)$', fontsize=16)

# fix the spacing between the subplots so they don't overlap
fig.tight_layout(pad=2.0)
````

For small values of $y$, we see that the function $g$ is not bounded by the blue line. 

For large enough values of $y$ (i.e. the limiting behaviour) we see that $g$ is bounded by a multiple of $y^4$.

Repeat our qualitative analysis of the discrete logistic map for the [Ricker map](https://en.wikipedia.org/wiki/Ricker_model), another classic model for population growth given by:
 
$$x_{n+1} = \exp(r(1-x_n))x_n$$

i.e. implement and plot its behaviour for different initial conditions and different $r$ parameters, and the bifurcation diagram.

### Solution - Ricker model

````python
def ricker_map(r, x):
    return np.exp(r*(1-x))*x

# plot Ricker map for parameter r = 1.5
r = 1.5
x = np.linspace(0, 3, 100)
y = ricker_map(r, x)

fig = plt.figure(figsize=(8, 6))
ax1 = fig.add_subplot(111)

ax1.plot(x,y,'b',label=r'$y = rx(1-x)$')
ax1.set_title('The ricker map ($r={0:.2f}$)'.format(r), fontsize=16)
ax1.set_xlabel('$x$', fontsize=14)
# add a grid
ax1.grid(True)

# plot fixed points
ax1.plot(x,x,'k',label=r'$y = x$')
ax1.plot(0, ricker_map(r,0), 'ro')
ax1.plot(1., ricker_map(r,1.), 'ro', label='intersection points')
ax1.legend(loc='best', fontsize=14)
````

````python
def plot_ricker_map_convergence(r, x0, iterations=10, ax=None):
    """Function to plot the iterations that the ricker map undergoes
    on axes ax (if given)

    Parameters
    ----------
    r : float
        Growth rate parameter
    x0 : float
        Starting population
    iterations : float, optional
        The number of iterations to compute
    ax : matplotlib axes, optional
        The axes to plot on
    """
    # start by plotting the underlying curves from above
    x = np.linspace(0, 3, 100)
    y = ricker_map(r, x)
    ax.plot(x,y,'b',label=r'$y = exp(r(1-x))x$')
    ax.plot(x,x,'k',label=r'$y = x$')
    ax.set_title('The ricker map ($r={0:.2f}$)'.format(r), fontsize=16)
    ax.set_xlabel('$x$', fontsize=14)
    ax.grid(True)
    ax.legend(loc='best', fontsize=14)
    # plot our starting point
    ax.plot(x0, x0, 'go') 
    # now iterate upto "iterations" and plot the path taken in (x,y) space
    x = x0
    for it in range(iterations):
        y = ricker_map(r,x)
        # plot some lines to indicate map progression
        ax.plot([x,x],[x,y],'k--',linewidth=1)
        ax.plot([x,y],[y,y],'k--',linewidth=1)
        # add a dot of increasing boldness
        ax.plot(x, y, 'ro', alpha=(it+1)/iterations)
        x=y
````

````python
# Let's have two plots side by side
fig, axs = plt.subplots(1, 2, figsize=(14, 7))

r = 1.9
x0 = 0.7
plot_ricker_map_convergence(r, x0, 10, axs[0])
x0 = 2
plot_ricker_map_convergence(r, x0, 10, axs[1])
````

Better yet, it has an [analytical solution](http://mathworld.wolfram.com/Analytic.html) and hence also provides an excellent case for us to test our numerical methods against.

Consider the following model problem which we will see *relaxes* $C$ to a value $C_r$:

$$\frac{dC}{dt} = K(C_r - C),$$

where $K$ is a coefficient that dictates how quickly the relaxation occurs, and $C_r$ is the value which we are relaxing to (can you see what the units of $K$ must be?).

This is a first-order ODE, and so integrating it up w.r.t. time once leads to a single arbitrary [constant of integration](https://en.wikipedia.org/wiki/Constant_of_integration). 

To arrive at a unique solution we thus need a single piece of auxiliary data - e.g. an initial condition which will generally take the form of specifying the value of the unknown at the initial time: $C(0)$.

This problem has the exact (analytical) solution

$$ C(t) = C_r + (C(0) - C_r)\exp(-Kt).$$

It is evident, both from the ODE and the solution, that

$$ \lim_{t\rightarrow\infty} C(t) = C_r.$$

As an example, suppose $C_r$ is zero, then this problem can be used to describe radioactive decay where $K$ is the inverse half-life (should give you a further clue on what the units of $K$ are!).

Below is an example of decay to zero using the half life of carbon-14, as used in [radiocarbon dating](https://en.wikipedia.org/wiki/Radiocarbon_dating).

````python
Cr = 0.0
C0 = 1.0
C_14_half_life = 5730. #  our temporal units here are years
# and the following is how you turn a "half life value" into a "decay constant"
K = np.log(2.)/C_14_half_life 

# let's plot things over 100,000 years
t = np.arange(0., 100000., 1.)

def Cex(t):
    """Function to evaluate the exact solution to the exponential decay problem.

    Parameters
    ----------
    t : numpy.ndarray or float
        Times

    Returns
    -------
    numpy.ndarray or float
        Function values for times given
    """ 
    return (Cr + (C0-Cr)*np.exp(-K*t))

fig = plt.figure(figsize=(6, 6))
ax1 = plt.subplot(111)
ax1.plot(t, Cex(t), 'b')

# red dashed lines and dot emphasise the meaning of "half life"
ax1.plot([5730], [0.5], 'ro')
ax1.plot([0, 5730, 5730], [0.5, 0.5, 0], 'r--')

ax1.set_xlabel(r'$t$ [years]', fontsize = 14)
ax1.set_ylabel(r'$C$', fontsize = 14)
ax1.set_title('Radioactive decay - exact solution', fontsize = 16)
ax1.grid(True)
ax1.set_xlim(0,100000)
ax1.set_ylim(0,1.1);
````

### Solution - Exponential decay (or relaxation)

````python
# set some physical parameters
K = 0.04
Cr = 20
C0 = 5

# set some numerical (time-stepping) parameters
dt = 11
t0 = 0
tmax = 200


# now i'm going to implement a function to perform FE integration for a general function f
def forward_euler(f, y0, t0, t_max, dt):
    """Forward Euler time stepper

    Parameters
    ----------
    f : function
        RHS function expected to be of the form f=f(t,y)
    y0 : float
        Initial y value
    t0 : float
        Initial time
    t_max : float
        The last timestep
    dt : float
        Timestep size

    Returns
    -------
    tuple of ndarray
        Tuple containing numpy arrays with all y and t values within
        the specified interval
    """
    y = np.array(y0)
    t = np.array(t0)
    y_all = [y0]
    t_all = [t0]
    while t < t_max:
        y = y + dt*f(t, y)  # euler guess
        y_all.append(y)
        t = t + dt
        t_all.append(t)
    return np.array(y_all), np.array(t_all)

def f(t, C):
    val = K * (Cr - C)
    return val

C_all, t_all = forward_euler(f, C0, t0, tmax, dt)



# now plot the solution
fig = plt.figure(figsize=(6, 6))
ax1 = plt.subplot(111)
ax1.plot(t_all, C_all, 'b.-', label='forward Euler')

def Cex(t):
    """Function to evaluate the exact solution to the exponential decay problem.

    Parameters
    ----------
    t : numpy.ndarray or float
        Times

    Returns
    -------
    numpy.ndarray or float
        Function values for times given
    """ 
    return (Cr + (C0-Cr)*np.exp(-K*t))

# let's use a finer resolution to plot the exact solution to give a smoother line.
t_fine = np.arange(0, tmax, dt/100)
ax1.plot(t_fine, Cex(t_fine), 'k', label='exact solution')
ax1.set_xlabel(r'$t$', fontsize = 16)
ax1.set_ylabel(r'$C$', fontsize = 16)
ax1.set_title('Numerical vs exact solution', fontsize = 16)
ax1.grid(True)
ax1.legend(loc='best', fontsize = 14);
````

So it's doing roughly the right thing; errors grow initially but for this particular problem they then reduce as both the numerical and exact solution converge (or asymptote) to the same value with time - this won't generally be the case and more normally we will just see errors keep growing.

You can try with different `dt` values to check qualitatively that the errors get smaller as you reduce `dt`.

We can formalise this with a convergence analysis:

[NB. to do this properly it's important to ensure that you always compare all the different numerical solutions with the exact solution consistently. Here we just compute the error at the final time and so for consistency we need to ensure that our algorithm always terminates at the same time. With floating point comparisons it's easy to mess this up.]

````python
# set some physical parameters
K = 0.04
Cr = 20
C0 = 5

# set some numerical (time-stepping) parameters
t0 = 0
tf = 40

# set up figure
fig, ax1 = plt.subplots(1, 1, figsize=(8, 5))

dt = 1
uall, tall = forward_euler(f, C0, t0, tf, dt)
ax1.plot(tall, uall, 'b', label='Forward Euler')

ax1.plot(tall, Cex(tall), 'k--', label='Exact')
ax1.set_xlabel('$t$', fontsize=14)
ax1.set_ylabel('$u(t)$', fontsize=14)
ax1.set_title('Exact vs numerical solutions, with a time step of {0:.5f}'.format(dt), fontsize=16)
ax1.legend(loc='best', fontsize=14)

# define function to compute error
def approx_error(dt, t_max):
    uall, tall = forward_euler(f, C0, t0, t_max, dt)
    err1 = abs(uall[-1]-Cex(tall[-1]))
    return err1

# compute error at each step
error_fe = []
dt_array = []
dt = 1
while dt > 1.e-4:
    dt_array.append(dt)
    err1 = approx_error(dt, tf)
    error_fe.append(err1)
    dt *= 0.5

# make convergence plot
fig, ax1 = plt.subplots(1, 1, figsize=(8, 5))
ax1.loglog(dt_array, error_fe, 'k.', label='Forward Euler')
ax1.set_xlabel('$\Delta t$', fontsize=14)
ax1.set_ylabel('Error at $t=3$', fontsize=14)
ax1.set_title('Convergence plot for solution at $t=40$', fontsize=16)
ax1.grid(True)
annotation.slope_marker((2e-3, 1.e-4), (1, 1), ax=ax1, size_frac=0.15, pad_frac=0.05)

# use numpy.polyfit to find best fit linear line to data
start_fit = 4  # what does this parameter do, and what happens if you choose a smaller value?
line_fit_fe = np.polyfit(np.log(dt_array[start_fit:]), np.log(error_fe[start_fit:]), 1)
ax1.loglog(dt_array, np.exp(line_fit_fe[1]) * dt_array**(line_fit_fe[0]), 'k-', label = 'slope: {:.2f}'.format(line_fit_fe[0]))

ax1.legend(loc='best', fontsize=14);
````

We'll come back to this in later lectures, but the slope here being 1 just means that the error decreases linearly with the time step size: halving `dt` aproximately halves the error.

We can, and in general will want to do better. Before we learn how to do this ourselves let's use an in-built Python (SciPy) ODE solver:

````python
# let's compare our forward Euler solver and results to scipy's solve_ivp functionality
from scipy.integrate import solve_ivp

# set some physical parameters
K = 0.04
Cr = 20
C0 = 5

# set some numerical (time-stepping) parameters
t0 = 0
tf = 40

def f(t, C):
    val = K * (Cr - C)
    return val

def Cex(t):
    """Function to evaluate the exact solution to the exponential decay problem.

    Parameters
    ----------
    t : numpy.ndarray or float
        Times

    Returns
    -------
    numpy.ndarray or float
        Function values for times given
    """ 
    return (Cr + (C0-Cr)*np.exp(-K*t))


# set up figure
fig, ax1 = plt.subplots(1, 1, figsize=(10, 5))

dt = 0.01
uall, tall = forward_euler(f, C0, t0, tf, dt)
ax1.semilogy(tall, np.abs(uall - Cex(tall)), 'b', label='Forward Euler')

# these are the OUTPUT times we want odeint to tell us the solution at
ts = np.linspace(t0, tf, 100)

# A solver tolerance - you can play with this value to see effect in figure below
tol = 1.e-5
# call the solver
sol = solve_ivp(f, [t0, tf], (C0,), method='RK45', t_eval=ts, rtol=tol, atol=tol)


ax1.plot(ts, np.abs(sol.y[0,:] - Cex(ts)), 'r', label='solve_IVP (RK45)')

ax1.set_xlabel('$t$', fontsize=14)
ax1.set_ylabel('Error', fontsize=14)
ax1.set_title('Error as a function of time', fontsize=16)
ax1.legend(loc='best', fontsize=14);
````

So note that the solve_IVP solver has errors several orders of magnitude smaller than the simple forward Euler method we implemented ourselves (for the `dt` value selected).

### Solution - Fibonacci numbers

````python
def compute_fib_number(n):
    """Function to compute the nth Fibonacci number.

    Parameters
    ----------
    n : int
        The index of the desired Fibonacci number

    Returns
    -------
    F_i_plus_1 : int
        The nth Fibonacci number F_n
    """
    F_i_minus_1 = 0
    if n==0: return 0
    F_i = 1
    if n==1: return 1
    F_i_plus_1 = 0
    for i in range(n-1):
        F_i_plus_1 =  F_i + F_i_minus_1
        F_i_minus_1 = F_i
        F_i = F_i_plus_1 
    return F_i_plus_1

print([compute_fib_number(i) for i in range(20)])
````
